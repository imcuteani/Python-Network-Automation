```python
import networkx as nx 
import matplotlib.pyplot as plt 
```

```python
# 1. Initiate an empty network graph 
G = nx.Graph()

# 2. Add network graph nodes and edges 
G.add_node("router-1")
G.add_nodes_from(["Cisco", "Cisco-IOS", "Aristra"])
G.add_edge("router-1", "switch-1")
G.add_edges_from([("Cisco", "Cisco-IOS"), ("Cisco-IOS", "Aristra"), ("Cisco", "Aristra")])
```

```python
# 3. Analyse the network with graph for SDN
shortest_path = nx.shortest_path(G, source="Cisco", target="Cisco-IOS")
print(f"Shortest path from Cisco catalyst to cisco IOS router is: {shortest_path}")  
```

```python
# 4. Visualize the SDN network graph 
nx.draw(G, with_labels=True, node_color='lightblue', font_weight='bold')
plt.show()
```
